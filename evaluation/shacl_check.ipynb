```python
from rdflib import Graph
from pyshacl import validate

# Load Knowledge Graph
data_graph = Graph()
data_graph.parse("scenarios_kgs.ttl", format="turtle")

# Load SHACL Shapes
shapes_graph = Graph()
shapes_graph.parse("shacl-rules.ttl", format="turtle")
```

```python
conforms, results_graph, results_text = validate(
    data_graph,
    shacl_graph=shapes_graph,
    inference='rdfs',          # or 'owlrl' if you want OWL RL reasoning
    abort_on_first=False,
    allow_warnings=True,
    meta_shacl=False,
    advanced=True
)
```
